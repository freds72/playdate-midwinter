```python
import math
import random
from PIL import Image,ImageDraw

mask = Image.open('./card_text_mask.png')
mask = mask.convert('L')

background = Image.open('./card_background.png')
background = background.convert('RGBA')

w,h = mask.size
t = 50
images = []
for k in range(32):
  snow = Image.new('RGBA',mask.size,color=(0,0,0,255))
  draw = ImageDraw.Draw(snow)
  random.seed(13)
  t += 0.2
  for i in range(1024):
    a = random.random() * math.pi * 2
    s = 2 + random.random()
    u = s * math.cos(a)
    v = s * abs(4*math.sin(a))
    x0 = random.randrange(0,w)
    x = int(x0 + t*u)%w
    y = int(86 + t*v)%h
    draw.rectangle(((x,y),(x+1,y+1)),(255,255,255,255))

  im = Image.composite(background, snow, mask)
  images.append(im)

images[0].save('./card_intro.gif', save_all=True, append_images=images[1:], optimize=False, duration=30, loop=0)

for i,img in enumerate(images):
  img.save(f'../source/system/card-highlighted/{i+1}.png',bits=1,optimize=False)

# display(images)
```

```python
import math
import random
from PIL import Image, ImageDraw, ImageFont

background = Image.open('../source/system/launchImage.png')
background = background.convert('RGBA')

# open font
font = ImageFont.truetype("Memo.ttf", 16)

text = ["loading","loading.","loading..","loading..."]
for i,t in enumerate(text):
  img = background.copy()
  draw = ImageDraw.Draw(img)
  draw.fontmode = '1'
  draw.text((337, 220), t, font=font, fill=(0,0,0,255))

  img.save(f'../source/system/launchImages/{i+1}.png',bits=1,optimize=False)
```
