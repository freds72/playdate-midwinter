```python
import math
import random
from PIL import Image,ImageDraw

mask = Image.open('./card_text_mask.png')
mask = mask.convert('L')

background = Image.open('./card_background.png')
background = background.convert('RGBA')

w,h = mask.size
h = 141-65
images = []
N = 60
for k in range(N):
  snow = Image.new('RGBA',mask.size,color=(0,0,0,255))
  draw = ImageDraw.Draw(snow)
  random.seed(13)
  t = k/N
  for i in range(796):
    a = math.pi * (2 + 2*t) + 4*random.random()
    s = 2 + 2 * random.random()
    u = 2*s * math.cos(a)
    v = s * abs(math.sin(a))
    x = int(random.randrange(0,w) + u)%w
    y = 32 + int(random.randrange(0,h) + h*t + v)%h
    r = random.random()>0.5 and 1 or 0
    draw.rectangle(((x,y),(x+r,y+r)),(255,255,255,255))

  im = Image.composite(background, snow, mask)
  images.append(im)

# images[0].save('./card_intro.gif', save_all=True, append_images=images[1:], optimize=False, duration=30, loop=0)

for i,img in enumerate(images):
  img.save(f'../source/system/card-highlighted/{i+1}.png',bits=1,optimize=False)

# display(images)
```

```python
import math
import random
from PIL import Image,ImageDraw

mask = Image.open('./icon_mask.png')
mask = mask.convert('L')

background = Image.open('./icon_background.png')
background = background.convert('RGBA')

w,h = mask.size
t = 50
images = []
for k in range(32):
  snow = Image.new('RGBA',mask.size,color=(0,0,0,255))
  draw = ImageDraw.Draw(snow)
  random.seed(13)
  t += 0.1
  for i in range(16):
    a = random.random() * math.pi * 2
    s = 2 + random.random()
    u = s * math.cos(a)
    v = s * abs(2*math.sin(a))
    x0 = random.randrange(0,w)
    x = int(x0 + t*u)%w
    y = int(86 + t*v)%h
    draw.rectangle(((x,y),(x+1,y+1)),(255,255,255,255))

  im = Image.composite(background, snow, mask)
  images.append(im)

# images[0].save('./icon.gif', save_all=True, append_images=images[1:], optimize=False, duration=30, loop=0)

for i,img in enumerate(images):
  img.save(f'../source/system/icon-highlighted/{i+1}.png',bits=1,optimize=False)
```

```python
import math
import random
from PIL import Image, ImageDraw, ImageFont

launch_card = Image.open('../source/system/card.png') 
launch_card = launch_card.convert('RGBA')
background = Image.open('./game_background.png')
background = background.convert('RGBA')

# img = Image.new('RGBA',background.size,color=(0,0,0,255))

def lerp(a,b,t):
	return a+(b-a)*t

images = []
for i in range(20):  
  img = Image.new('RGBA',background.size,color=(0,0,0,0))
  img.paste(launch_card, (25, 43))
  mask = Image.new("L", background.size, 0)
  # draw circles
  draw = ImageDraw.Draw(mask)
  random.seed(i)
  for x in range(0,400,16):
    for y in range(0,240,16):
      x0 = x + random.uniform(-8,8)
      y0 = y + random.uniform(-8,8)
      r = lerp(-16-y/8,0,x/400)
      r = max(0,r+58*i/20)
      leftUpPoint = (x0-r, y0-r)
      rightDownPoint = (x0+r, y0+r)
      twoPointList = [leftUpPoint, rightDownPoint]
      draw.ellipse(twoPointList, fill=255)			

      im = Image.composite(background, img, mask)
  images.append(im)
  im.save(f'../source/system/launchImages/{i+1}.png',bits=1,optimize=False)

images[0].save('./launch-card.gif', save_all=True, append_images=images[1:], optimize=False, duration=30, loop=0)  
```
