```python
data = [[5," "],[11,"%"],[8,"AMWmw_"],[7,"BDGHJKLNOPQRSTUVXYabdeghknopqrsuvxy0123456789?+-="],[6,"CEFZcftz!\"*"],[5,"Iij()"],[4,".,;:'"],[3,"l"]]

charset = "ABCDEFGHIJKLMNOPQRSTUVWXYZabcdefghijklmnopqrstuvwxyz0123456789.,;:?!\"'+-=*%_()"
spaces = {}
for category in data:
  space, chars = category
  for c in chars:
    spaces[c] = space

print("tracking=0")
for c in charset:
  ch = c
  if ch==" ": ch="space"
  print(ch,"\t",spaces[c])
```
